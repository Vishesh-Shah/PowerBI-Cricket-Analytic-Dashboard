## Batting Avg = total_run/total_dismissals

total_dismmials = sum of the out

```python
def batting_avg(df1):
    df2= df1.copy()
    dic = {}
    for i in df2['batsmanName']:
        batting_avg = df2[df2['batsmanName'] == i]['runs'].sum()/df2[df2['batsmanName']==i]['out'].sum()
        if batting_avg =='nan':
            dic[i] = 0
        else:
            dic[i] = np.round(batting_avg,2)
    print(dic['Naveen-ul-Haq'])
    df2['Batting_Average'] = df2['batsmanName'].map(dic)
    return df2
```

```python
bat_df2 = batting_avg(bat_df1)
```

# Strike_Rate = Toal_runs/total_ball_face

```python
def strike_rate(df1):
    df2= df1.copy()
    dic = {}
    for i in df2['batsmanName']:
        strike_rate = (df2[df2['batsmanName'] == i]['runs'].sum()/df2[df2['batsmanName']==i]['balls'].sum())*100
        if strike_rate =='nan':
            dic[i] = 0
        else:
            dic[i] = np.round(strike_rate,2)
    print(dic['Naveen-ul-Haq'])
    df2['Strike_Rate'] = df2['batsmanName'].map(dic)
    return df2
```

```python
bat_df3 = strike_rate(bat_df2)
bat_df3
```
